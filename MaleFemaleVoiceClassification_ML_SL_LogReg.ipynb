# Male Female voice classification
### From "voice.csv" dataset, use acoustic features to distinguish a male from female.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn import preprocessing
```

```python
df = pd.read_csv('voice.csv')
```

```python
df.head()
```

#### We will use acoustic features to distinguish a male voice from female. Load the dataset from “voice.csv”, identify the target variable and do a one-hot encoding for the same. Split the dataset in train-test with 20% of the data kept aside for testing.

```python
# label is our target variable. Using one-hot encoder to transform it
df.label = preprocessing.LabelEncoder().fit_transform(df.label)
```

```python
x = df.loc[:,:'modindx']
y = df.label
```

```python
# test size 20%
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=.2,random_state=4)
```

#### Fitting a logistic regression model and measure the accuracy on the test set.

```python
logistic = LogisticRegression()
logistic.fit(x_train,y_train)
```

```python
pred_y = logistic.predict(x_test)
metrics.accuracy_score(pred_y, y_test)
```

#### Computing the correlation matrix that describes the dependence between all predictors, identifying the predictors that are highly correlated and plotting the correlation matrix using seaborn heatmap.

```python
x.corr()
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(x.corr(),cbar=True,square=True,cmap='coolwarm')
plt.show()
```

```python
# get all strong correlations (>= 0.75)
import itertools
dfp = pd.DataFrame([[(i,j),x.corr().loc[i,j]] for i,j in list(itertools.combinations(x.corr(), 2))],columns=['pairs','corr'])    
dfp.sort_values(by='corr',ascending=False).head(15)
```

```python
# Highly correlated columns:
# ['meanfreq','centroid','maxdom','dfrange','skew','kurt','median','Q25','sd','IQR','sfm','meandom','sp.ent','median']
```

#### Removing those predictors that are correlated and fit a logistic regression model again and comparing the accuracy with that of previous model

```python
new_x = df[['Q75','mode','meanfun','minfun','maxfun','mindom','modindx']]
```

```python
# train
x_train, x_test, y_train, y_test = train_test_split(new_x,y,test_size=.2,random_state=4)
```

```python
# fit
logistic = LogisticRegression()
logistic.fit(x_train,y_train)
```

```python
# predict and accuracy
pred_y = logistic.predict(x_test)
metrics.accuracy_score(pred_y, y_test)
```
